1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

```python
'''
The different types of data are:

1. Quantitative Data : Numerical type of data that can be measured are called Quantitative data.
    eg :  Number of students in a class(20,30,40)
         Height of people(165.5 cm, 170.2 cm)
    They are of two types
      i. Discrete Data          and        ii. Continuous Data
2. Qualitative Data : Non-numerical or categorical data or attributes that cannot be measured but counted, are called Qualitative data.
    eg : Colors(Red,Blue,Green)
        Rankings(1st, 2nd,3rd)
    They are of two types
      i. Nominal Data           and        ii. Ordinal Data

The measurement scales of data are of four types that are  nominal,  ordinal,  interval, and  ratio scales.

Nominal Scale : It is a type of scale that labels or categorizes data without any quantitative value or order.
                  There is no inherent order or ranking.
                  eg: colors(Red, Blue, Green) ,
                   Gender(Male and Female),
                   Product categories(Electronic, Clothing).
Ordinal Scale : It is a type of scale that categorizes data with an inherent order or ranking,  but the differences
                between the ranks are not equal.
                 eg : Rankings(1st, 2nd, 3rd),
                  Ratings(Poor, Fair, Good, Excellent)
Interval Scale : It is a type of scale that measures data with equal intervals between consecutive levels, but lacks
                a true zero point.
                eg : Temperature(in celcius or fahrenheit)
                  IQ scores, etc.
Ratio Scale : It is a type of scale that measures data with equal intervals betweem consecutive levels and has a
                true zero starting point.
                eg : Weight, Height, Age, Income ,etc.
'''
```

2. What are the measures of central tendency, and when should you use each. Discuss the mean, median, and mode with examples and situations where each is appropriate.

```python
'''
The measures of the central tendency includes:

1. MEAN : The mean is the average value of the dataset. The formula of mean is given by:
                   mean= (sum of all observations / no.of observations)
           Mean is generally used in case of  normally distributed data (without outliers),
            for the data that is continuous and quantitative.
            and includes all the data points in the calculations.

2. MEDIAN : The median is the middle value of the dataset when it is ordered from smallest to largest.
              the median value of an odd dataset is the middle occuring value, and that od an even dataset,
               it is the average value of the two middle values.
              Median is mostly used in case of Skewed data or data having outliers, or when the data is
               ordinal or ranked.

3. MODE : The mode is the value that appears most frequently in the dataset. It is the value having the 4
                highest frequency.
              it is used mostly in case of categorical or nominal datasets. to identify the most common value.
              Data having multiple frequeny  peaks is called multimodal data.
'''
```

```python
# for example:

a= [2,3,3,4,5,5,6,6,6,7,7,8,9,9]

import numpy as np
import statistics

print("the mean is :",np.mean(a))

print("the median is :", np.median(a))

print("the mode of the data is :", statistics.mode(a))
```

3. Explain the concept of dispersion. How do variance and standard deviation measures the spread of data ?


```python
'''
Dispersion refers to the extent to which data points in Datasets deviate from the central tendency(i.e. mean,median,or mode.)
It is the measure of the spred of the data.
The measures of dispersion are : Range, IQR, Variance, and Standard Deviation.

VARIANCE :
 The average of the squared difference between each data point and the mean, is called variance.
  it is denoted by (sigma)^2.
  Higher the value of variance denotes the greater spread.

STANDARD DEVIATION :
 The square root of variance is called standard variation,
  it is denoted by sigma.
  (It is also called the root mean square value of the data)
'''
```

6. Explain the concept of skewness and its types. How does skewness effect the interpretation of data ?

```python
'''
Skewness is the measure of the asymmetry of the probability distribution of the real-valued random variable.
it indicates whether the data is symmetric or skewed to one side .

There are three types of skewness :
1. Positive Skewness : also called Right Skewed . The tail on the right side of the distribution is longer
     or fatter than the left side. The mean is greater than the median in this case.

2. Negative Skewness : also called Left Skewed . The tail on the left side of the distribution is longer
     or fatter than the right side. The mean is smaller than the median in this case.

3. Zero Skewness : Is called the symmetric data. The distribution in this case is symmetric and equal on both the sides.
      The mean , median, and the mode are all equal.

Skewness affects the interpretation of data as :
    i.  It affect the relation between the mean and the median, the mean is towards the skew.
   ii.  Skewness can impact the representation of the data, making it essential to consider the type of skewness
           when selecting the statistical methods.
  iii.  Skewness can be influenced by the outliers, which can affect the interpretation of the data.
'''
```

7. What is the interquartile region(IQR), and how is it used to detect outliers ?

```python
'''
The interquartile region is the region between the 1st and the 3rd quartile .
IQR is a measure of statistical dispersion, specifically representing the range of the middle 50% of a dataset.
 It's calculated by subtracting the first quartile (Q1) from the third quartile (Q3) of a dataset.

    IQR = Q3 - Q1

to detect the outliers , we first find out the upper and the lower fence by the formula:
        Lower Fence: Q1 - 1.5 * IQR
        Upper Fence: Q3 + 1.5 * IQR
Any data point that falls below the lower fence or above the upper fence is considered a potential outlier.
'''
```

9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).

```python
'''
A normal distribution is a bell-shaped, symmetrical probability distribution.
The mean, median, and mode are equal in a normal distribution. And it is visuallised as a Bell- shaped curve.
The total area under the curve is equal to 1 , and it extends infinitely in both directions, approaching the x-axis but never touching it.

The Empirical rule :
The empirical rule describes the percentage of data within certain standard deviations of the mean in a normal distribution.

(68-95-99.7 rule) states that :
   Approximately 68% of data falls within 1 standard deviation of the mean.
   Approximately 95% of data falls within 2 standard deviations of the mean. and
   Approximately 99.7% of data falls within 3 standard deviations of the mean.
'''
```

10. Provide a real-life example of a Poisson process and calculate the probability for a specific event.

```python
# Poisson Process :

# eg : the number of customers arriving at a grocery store per hour, assuming an average arrival rate.

from scipy.stats import poisson
import numpy as np

avg_arrival_rate = 10
X = 15

probability = poisson.pmf(X, avg_arrival_rate)

print(f"The probability of {X} customers arriving is: {probability:.4f}")
```

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

```python
'''
Random Variable can be defined as a variable, whose numerical value is determined by the outcome of a random experiment.
 It can be discrete, taking on specific integral values, or continuous, taking on any value within a range.

Discrete Random Variable :
   A discrete random variable can only take on specific, separate values, often whole numbers.
   These values can be counted, and there are gaps or no values between them.
   the probability distribution of a discrete random variable is defined by Probability mass function (PMF).
      for eg :
          the number of heads in multiple coin flips,
          the number of cars passing a certain point, or
          the number of defective items in a batch.

Continuous Random Variable :
   A continuous random variable can take on any value within a given range or interval.
   There are infinitely many possible values between any two given values.
   the probability distribution of a continuous random variable is defined by Probability density function (PDF).
      for eg :
       height, weight, temperature, or
       the time taken to complete a task.
'''
```
