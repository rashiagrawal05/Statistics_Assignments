16. How do you calculate a Z-score, and what does it represent ?

```python
'''
Z-score is also known as a standard score.
It measures how many standard deviations a data point is from the mean of a distribution.
It is calculated by subtracting the mean from the data point and then dividing the result by the standard deviation.
i.e.
      z = (x - μ) / σ

      where, z is the z-score,
             x is the data point,
             μ is the population mean, and
            σ is the population standard deviation.

A positive z-score indicates the data point is above the mean,
and  a negative z-score indicates it is below the mean.
'''
```

22.  What is the concept of expected value in a probability distribution ?

```python
'''
The expected value in a probability distribution represents the long-run average outcome of a random variable.
For a discrete random variable, the expected value is calculated by multiplying each possible outcome by its probability
 and summing the results.

     E(X) = Σ [x * P(x)]

           where 'x' represents each outcome
           and 'P(x)' is its probability.
'''
```

14.  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

```python
def calculate_z_score(data_point, mean, std_dev):
  if std_dev == 0:
    raise ValueError("Standard deviation cannot be zero.")
  z_score = (data_point - mean) / std_dev
  return z_score

def compare_to_standard_normal(z_score):
  cumulative_probability = norm.cdf(z_score)
  percentile = cumulative_probability * 100

  print(f"\nComparing Z-score to Standard Normal Distribution:")
  print(f"Cumulative probability (P(Z <= {z_score:.2f})): {cumulative_probability:.4f}")
  print(f"This Z-score corresponds to the {percentile:.2f}th percentile.")

  if z_score > 2 or z_score < -2:
    print("This Z-score suggests the data point is relatively far from the mean (more than 2 standard deviations).")
  elif z_score > 1 or z_score < -1:
    print("This Z-score suggests the data point is moderately far from the mean (between 1 and 2 standard deviations).")
  else:
    print("This Z-score suggests the data point is relatively close to the mean (within 1 standard deviation).")
```

```python
data = np.array([6, 7, 7, 12, 13, 13, 15, 16, 19, 22])
data_mean = np.mean(data)
data_std_dev = np.std(data)

print(f"Dataset: {data}")
print(f"Mean of dataset: {data_mean:.2f}")
print(f"Standard deviation of dataset: {data_std_dev:.2f}")

z_score = calculate_z_score(data,data_mean,data_std_dev)
print(z_score)
```

15.  Implement hypothesis testing using Z-statistics for a sample dataset.


```python
import scipy.stats as stats

def z_test(sample_mean, p_mean, p_std, sample_size=100, alpha=0.05, alternative='two-sided'):
  z_statistic = (sample_mean - p_mean) / (p_std / np.sqrt(sample_size))

  if alternative == 'two-sided':
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
  elif alternative == 'less':
    p_value = stats.norm.cdf(z_statistic)
  elif alternative == 'greater':
    p_value = 1 - stats.norm.cdf(z_statistic)
  else:
    raise ValueError("Invalid alternative. Choose 'two-sided', 'less', or 'greater'.")

  print(f"Z-statistic: {z_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")
  print(f"Significance level (alpha): {alpha}")

  if p_value < alpha:
    print("Decision: Reject the null hypothesis.")
  else:
    print("Decision: Fail to reject the null hypothesis.")

  return z_statistic, p_value
```

```python
sample_mean = 11.5
population_mean = 12
population_std_dev = 1.5
sample_size = 50
alpha = 0.05

print("Performing Z-test (two-sided):")
z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=alpha, alternative='two-sided')
```

```python
print("Performing Z-test (less than):")
z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=alpha, alternative='less')
```

16. Create a confidence interval for a dataset using Python and interpret the result.

```python
data = np.array([12, 15, 14, 10, 13, 17, 14, 15, 16, 14])

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)
confidence_level = 0.95

alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha / 2, df=n - 1)

standard_error = sample_std / np.sqrt(n)

margin_of_error = t_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Number of Samples: {n}")
print(f"Confidence Level: {confidence_level * 100}%")
print(f"T-critical value: {t_critical:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

23. Implement Z-score normalization on a dataset and visualize its transformation.

```python
def z_score_normalize(data):
  mean = np.mean(data)
  std_dev = np.std(data)
  z_scores = (data - mean) / std_dev
  return z_scores

data = np.array([15, 18, 20, 22, 25, 28, 30, 32, 35, 38, 40])

normalized_data = z_score_normalize(data)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=5, color='skyblue', edgecolor='black')
plt.title('Original Data Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=5, color='lightcoral', edgecolor='black')
plt.title('Z-Score Normalized Data Distribution')
plt.xlabel('Z-scores')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
