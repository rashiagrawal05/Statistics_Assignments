1.  What is hypothesis testing in statistics ?

```python
'''
Hypothesis testing in statistics is a method used to assess a claim about a population parameter using sample data.
 It involves formulating a null hypothesis (a statement of no effect or no difference) and
 an alternative hypothesis (a statement of what the researcher believes to be true).
'''
```

2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?

```python
'''
Null Hypothesis is the initial assumption, often stating no effect, no difference, or no relationship.
Alternative Hypothesis is the researcher's claim, contradicting the null hypothesis.
'''
```

3.  What is the significance level in hypothesis testing, and why is it important ?

```python
'''
The significance level in hypothesis testing is a critical component of hypothesis testing in statistics.
It represents the maximum probability of rejecting the null hypothesis when it is actually true.

It is important because it ensures research reliability, Controls the errors, Increases confidence in results,
and provides a standard framework for evaluating evidences and decision making about the hypothesis.
'''
```

4. What does a P-value represent in hypothesis testing ?

```python
'''
The P-value in hypothesis testing represents the probability of observing a test statistic atleast as extreme
 as the one observed, assuming that the null hypothesis is true.
'''
```

5.  How do you interpret the P-value in hypothesis testing ?

```python
'''
The P-value interpretation :

    1. Small p-value : Small p-value (< 0.05) indicates strong evidence against the null hypothesis,
suggesting that the observed results are unlikely to occur by chance.

    2. Large p-value : A Large p-value (> 0.05) indicates a weak evidence against the null hypothesis,
suggesting that the observed results maybe occur by chance.
'''
```

6.  What are Type 1 and Type 2 errors in hypothesis testing ?

```python
'''
In hypothesis testing, there are two types of errors:

Type1 error :  Rejecting a Null Hypothesis (H0) which is actually true, is called Type1 error.
     eg :  Convicting an innocent person.

Type2 errror :  Failing to reject a Null Hypothesis (H0) which is actually false, is called as Type2 error.
     eg :  Aquitting a guilty person.

'''
```

7.  What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?

```python
'''
One - Tailed test :
      Tests for an effect in one specific direction. (eg., greater than or less than).
    The alternate hypothesis states a specific direction of the effect, (H1 : mu > mu0  or H1 : mu < mu0).
    The critical region is located in the one tail of the  distribution.

Two - Tailed test :
      Tests for an effect in both the direction. (eg., equal or not equal to).
    The alternate hypothesis does not specify the direction of the effect, (H1 : mu != mu0).
    The critical region is divided between both the tails of the  distribution.
'''
```

1.  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

```python
import math
import scipy.stats as st

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
  z_statistic = (sample_mean - population_mean) / (population_std / math.sqrt(sample_size))
  p_value = 2 * (1 - st.norm.cdf(abs(z_statistic)))
  reject_null = p_value < alpha
  return z_statistic, p_value, reject_null

def interpret_results(z_statistic, p_value, reject_null):
  print(f"Z-statistic: {z_statistic:.4f}")
  print(f"P-value: {p_value:.4f}")

  if reject_null:
    print("Reject the null hypothesis.")
    print("There is a statistically significant difference between the sample mean and the population mean.")
  else:
    print("Fail to reject the null hypothesis.")
    print("There is no statistically significant difference between the sample mean and the population mean.")
```

```python
sample_mean = 78
population_mean = 75
population_std = 10
sample_size = 40
alpha = 0.05
z_statistic, p_value, reject_null = z_test(sample_mean, population_mean, population_std, sample_size, alpha)
interpret_results(z_statistic, p_value, reject_null)
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

sample_mean = 52
hypothesized_mu = 50
population_std = 10
sample_size = 100
alpha = 0.05

standard_error = population_std / np.sqrt(sample_size)
z_statistic = (sample_mean - hypothesized_mu) / standard_error

critical_value_lower = stats.norm.ppf(alpha / 2)
critical_value_upper = stats.norm.ppf(1 - (alpha / 2))

print(f"Calculated Z-statistic: {z_statistic:.2f}")
print(f"Critical Z-values (lower, upper): {critical_value_lower:.2f}, {critical_value_upper:.2f}")

if z_statistic < critical_value_lower or z_statistic > critical_value_upper:
    print("Decision: Reject the Null Hypothesis (H0).")
    print("There is sufficient evidence to suggest the population mean is significantly different from the hypothesized mean.")
else:
    print("Decision: Fail to Reject the Null Hypothesis (H0).")
    print("There is not enough evidence to suggest the population mean is significantly different from the hypothesized mean.")
```

```python
x = np.linspace(-4, 4, 500)
pdf = stats.norm.pdf(x, 0, 1)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')

x_reject_lower = x[x < critical_value_lower]
pdf_reject_lower = pdf[x < critical_value_lower]
plt.fill_between(x_reject_lower, 0, pdf_reject_lower, color='red', alpha=0.3, label='Rejection Region (Lower)')

x_reject_upper = x[x > critical_value_upper]
pdf_reject_upper = pdf[x > critical_value_upper]
plt.fill_between(x_reject_upper, 0, pdf_reject_upper, color='red', alpha=0.3, label='Rejection Region (Upper)')

plt.axvline(z_statistic, color='green', linestyle='--', label=f'Calculated Z-statistic: {z_statistic:.2f}')

plt.axvline(critical_value_lower, color='purple', linestyle=':', label=f'Critical Z-value (Lower): {critical_value_lower:.2f}')
plt.axvline(critical_value_upper, color='purple', linestyle=':', label=f'Critical Z-value (Upper): {critical_value_upper:.2f}')

plt.title('Two-Tailed Z-Test: Decision Region Visualization')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def error_type(mu0,mu1,sigma,n,alpha):
  critical_value = norm.ppf(1-alpha,loc=mu0,scale=sigma/np.sqrt(n))

  #alpha
  type1_error = 1-norm.cdf(critical_value,loc=mu0,scale=sigma/np.sqrt(n))
  #beta
  type2_error = norm.cdf(critical_value,loc=mu1,scale=sigma/np.sqrt(n))

  power = 1-type2_error

  return type1_error,type2_error,power,critical_value
```

```python
#Visualisation

def visualise_errors(mu0,mu1,sigma,n,alpha,critical_value):
  x= np.linspace(mu0-3*sigma/np.sqrt(n), mu1+3*sigma/np.sqrt(n),100)

  plt.plot(x,norm.pdf(x,loc=mu0,scale=sigma/np.sqrt(n)),label= "Null Hypothesis")
  plt.plot(x,norm.pdf(x,loc=mu1,scale=sigma/np.sqrt(n)),label = "Alternate Hypothesis")

  plt.axvline(critical_value,color = "r",label="Critical Value")
  plt.fill_between(x,norm.pdf(x,loc=mu0,scale=sigma/np.sqrt(n)),where=(x>critical_value),alpha=0.3,label='Type1 Error')
  plt.fill_between(x,norm.pdf(x,loc=mu1,scale=sigma/np.sqrt(n)),where=(x>critical_value),alpha=0.3,label='Type1 Error')

  plt.legend()
  plt.show()
```

```python
#example

mu0 = 0
mu1 = 1
sigma = 1
n=100
alpha = 0.05

type1_error,type2_error,power,critical_value = error_type(mu0,mu1,sigma,n,alpha)
print("Type1 error : ",type1_error)
print("Type2 error : ",type2_error)
print("POWER : ",power)
print('Critical Value : ',critical_value)
```

```python
visualise_errors(mu0,mu1,sigma,n,alpha,critical_value)
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def prior(theta):
  return norm.pdf(theta,loc=0,scale=1)

def likelihood(data,theta):
  return norm.pdf(data,loc=theta)

def posterior(data,theta):
  likelihood_values = likelihood(data,theta)
  posterior_value = np.sum(np.log(likelihood_values)) + np.log(prior(theta))
  return posterior_value

np.random.seed(0)
theta_true = 2
data = np.random.normal(loc = theta_true,size=100)

theta_values = np.linspace(-5,5,100)
posterior_value = np.array([posterior(data,theta) for theta in theta_values])

posterior_probabilities = np.exp(posterior_value-np.max(posterior_value))
posterior_probabilities /= np.sum(posterior_probabilities)
```

```python
plt.plot(theta_values,posterior_probabilities)
plt.title("Posterior Distribution")
plt.xlabel('Theta')
plt.ylabel("Posterior Probability")
plt.show()

map_estimate = theta_values[np.argmax(posterior_probabilities)]
print(f"MAP Estimate : {map_estimate}")
```

16. Implement an F-test using Python to compare the variances of two random samples.

```python
import numpy as np
from scipy.stats import f

sample1 = np.random.normal(loc=0,scale=1,size=100)
sample2 = np.random.normal(loc=0,scale=2,size=100)

var1 = np.var(sample1,ddof=1)
var2 = np.var(sample2,ddof=1)

f_stat = var1/var2

df1 = len(sample1)-1
df2 = len(sample2)-1

p_value = 2*(1- f.cdf(f_stat,df1,df2))

print("F-statistic : ",f_stat)
print("P-value : ",p_value)

alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis. The variances are likely different.")
else:
  print("Failed to reject the null hypothesis. The variances are likely equal.")
```

24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

```python
import numpy as np
import scipy.stats as stats

def chi_square_variance_test(sample_data, hypothesized_variance, alpha=0.05):
  sample_size = len(sample_data)
  sample_variance = np.var(sample_data, ddof=1)  # Use ddof=1 for sample variance

  chi_squared_stat = (sample_size - 1) * sample_variance / hypothesized_variance
  degrees_of_freedom = sample_size - 1
  p_value = 1 - stats.chi2.cdf(chi_squared_stat, degrees_of_freedom)

  critical_value = stats.chi2.ppf(1 - alpha, degrees_of_freedom)

  if chi_squared_stat > critical_value:
    decision = "Reject the null hypothesis"
  else:
    decision = "Fail to reject the null hypothesis"

  return chi_squared_stat, p_value, critical_value, decision
```

```python
# Example usage
sample_data = [12, 15, 18, 20, 22, 25, 28, 30, 32, 35]
hypothesized_variance = 50
alpha = 0.05

test_statistic, p_value, critical_value, decision = chi_square_variance_test(sample_data, hypothesized_variance, alpha)

print(f"Chi-square statistic: {test_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical value: {critical_value:.4f}")
print(f"Decision: {decision}")
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def two_proportion_z_test(successes1, observations1, successes2, observations2, alpha=0.05, alternative='two-sided'):
    count = np.array([successes1, successes2])
    nobs = np.array([observations1, observations2])

    z_statistic, p_value = proportions_ztest(count, nobs, alternative=alternative)

    if p_value < alpha:
        decision = "Reject the Null Hypothesis (H0): There is a significant difference in proportions."
    else:
        decision = "Fail to Reject the Null Hypothesis (H0): No significant difference in proportions."

    return z_statistic, p_value, decision

# Example Usage:
clicks_campaign_A = 120
total_views_campaign_A = 500

clicks_campaign_B = 90
total_views_campaign_B = 400

z_stat, p_val, test_decision = two_proportion_z_test(
    clicks_campaign_A, total_views_campaign_A,
    clicks_campaign_B, total_views_campaign_B
)

print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Decision: {test_decision}")
```

26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

worker1 = [18,34,17,20,56,23,12,56,12,35,57]
worker2 = [28,44,19,32,56,80,52,26,52,30,29]

f_stats = np.var(worker1)/np.var(worker2)
f_stats
```
